```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np
import xlsxwriter
sayfalar = [""]*33
Net_Metrekare = [""]*1020
Oda_Sayisi = [""]*1020
Isitma_Tipi = [""]*1020
Krediye_Uygunluk = [""]*1020
Binanin_Yasi = [""]*1020
Binanin_Kat_Sayisi = [""]*1020
fiyatlar = [""]*1020
ListIndex=0
```

```python
for x in range(1,35):
    if x==1:
       isYerleri = BeautifulSoup(requests.get("https://www.emlakjet.com/satilik-isyeri/ankara/?ilce[]=ankara-yenimahalle-1723&ilce[]=ankara-kecioren-1745&ilce[]=ankara-pursaklar-2034&ilce[]=ankara-mamak-1746&ilce[]=ankara-etimesgut-1922&ilce[]=ankara-cankaya-1231&ilce[]=ankara-altindag-1130").content,"lxml").find_all("div",attrs={"class":"_3qUI9q"})    
    else:
       sayfalar[x-2] += "https://www.emlakjet.com/satilik-isyeri/ankara/"+str(x)+"/?ilce[]=ankara-yenimahalle-1723&ilce[]=ankara-kecioren-1745&ilce[]=ankara-pursaklar-2034&ilce[]=ankara-mamak-1746&ilce[]=ankara-etimesgut-1922&ilce[]=ankara-cankaya-1231&ilce[]=ankara-altindag-1130" 
       isYerleri = BeautifulSoup(requests.get(sayfalar[x-2]).content,"lxml").find_all("div",attrs={"class":"_3qUI9q"})   
    for link in isYerleri:
        linkler = link.a.get("href")
        detaylar = BeautifulSoup(requests.get("https://www.emlakjet.com"+linkler).content,"lxml")
        isYeriDetaylari = detaylar.find_all("div",attrs={"class":"_35T4WV"})
        detaylar.find("div",attrs={"class":"ukxyhf"}).decompose()
        fiyatlar[ListIndex] += detaylar.find("div",attrs={"class":"_2TxNQv"}).text.replace(',', '')
        for detay in isYeriDetaylari:
            if detay.find("div",attrs={"class":"_1bVOdb"}).text=="Net Metrekare":  
               Net_Metrekare[ListIndex] += detay.find("div",attrs={"class":"_1bVOdb"}).next_sibling.text.upper().replace('M2', '')
            if detay.find("div",attrs={"class":"_1bVOdb"}).text=="Oda Sayısı":
               Oda_Sayisi[ListIndex] += detay.find("div",attrs={"class":"_1bVOdb"}).next_sibling.text.upper()[0]
            if detay.find("div",attrs={"class":"_1bVOdb"}).text=="Isıtma Tipi":
               Isitma_Tipi[ListIndex] += detay.find("div",attrs={"class":"_1bVOdb"}).next_sibling.text.upper()   
            if detay.find("div",attrs={"class":"_1bVOdb"}).text=="Krediye Uygunluk":
               Krediye_Uygunluk[ListIndex] += detay.find("div",attrs={"class":"_1bVOdb"}).next_sibling.text.upper()
            if detay.find("div",attrs={"class":"_1bVOdb"}).text=="Binanın Yaşı":
               Binanin_Yasi[ListIndex] += detay.find("div",attrs={"class":"_1bVOdb"}).next_sibling.text.upper()
            if detay.find("div",attrs={"class":"_1bVOdb"}).text=="Binanın Kat Sayısı":
               Binanin_Kat_Sayisi[ListIndex] += detay.find("div",attrs={"class":"_1bVOdb"}).next_sibling.text.upper()
        ListIndex +=1
```

```python
for i in range(0,1020):
    print(Net_Metrekare[i]+" --- "+Oda_Sayisi[i]+" --- "+Isitma_Tipi[i]+" --- "+Krediye_Uygunluk[i]+" --- "+Binanin_Yasi[i]+" --- "+Binanin_Kat_Sayisi[i]+" --- "+fiyatlar[i])

df = pd.DataFrame({
    "Net Metrekare": Net_Metrekare,
    "Oda Sayısı": Oda_Sayisi,
    "Isıtma Tipi": Isitma_Tipi,
    "Krediye Uygunluk": Krediye_Uygunluk,
    "Binanın Yaşı": Binanin_Yasi,
    "Binanın Kat Sayısı": Binanin_Kat_Sayisi,
    "Fiyat": fiyatlar   
})
df.to_excel("is_yeri_verileri02.xlsx", index=False)
```
