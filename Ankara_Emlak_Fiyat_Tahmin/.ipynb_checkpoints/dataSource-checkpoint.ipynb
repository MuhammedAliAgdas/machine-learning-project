```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np
import xlsxwriter
sayfalar = [""]*35
Net_Metrekare = [""]*1060
Oda_Sayisi = [""]*1060
Isitma_Tipi = [""]*1060
Krediye_Uygunluk = [""]*1060
Binanin_Yasi = [""]*1060
Binanin_Kat_Sayisi = [""]*1060
fiyatlar = [""]*1060
ListIndex=0
for x in range(1,37):  
    if x==1:
       arsalar = BeautifulSoup(requests.get("https://www.emlakjet.com/satilik-isyeri/ankara/?ilce[]=ankara-yenimahalle-1723&ilce[]=ankara-kecioren-1745&ilce[]=ankara-pursaklar-2034&ilce[]=ankara-mamak-1746&ilce[]=ankara-etimesgut-1922&ilce[]=ankara-cankaya-1231&ilce[]=ankara-altindag-1130").content,"lxml").find_all("div",attrs={"class":"_3qUI9q"})    
    else:
       sayfalar[x-2]+="https://www.emlakjet.com/satilik-isyeri/ankara/"+str(x)+"/?ilce[]=ankara-yenimahalle-1723&ilce[]=ankara-kecioren-1745&ilce[]=ankara-pursaklar-2034&ilce[]=ankara-mamak-1746&ilce[]=ankara-etimesgut-1922&ilce[]=ankara-cankaya-1231&ilce[]=ankara-altindag-1130" 
       arsalar = BeautifulSoup(requests.get(sayfalar[x-2]).content,"lxml").find_all("div",attrs={"class":"_3qUI9q"})   
    for link in arsalar:
        linkler = link.a.get("href")
        detaylar = BeautifulSoup(requests.get("https://www.emlakjet.com"+linkler).content,"lxml")
        arsaDetaylari = detaylar.find_all("div",attrs={"class":"_35T4WV"})
        detaylar.find("div",attrs={"class":"ukxyhf"}).decompose()
        fiyatlar[ListIndex] += detaylar.find("div",attrs={"class":"_2TxNQv"}).text
        for detay in arsaDetaylari:
            if detay.find("div",attrs={"class":"_1bVOdb"}).text=="Net Metrekare":  
               Net_Metrekare[ListIndex] += detay.find("div",attrs={"class":"_1bVOdb"}).next_sibling.text
            if detay.find("div",attrs={"class":"_1bVOdb"}).text=="Oda Sayısı":
               Oda_Sayisi[ListIndex] += detay.find("div",attrs={"class":"_1bVOdb"}).next_sibling.text
            if detay.find("div",attrs={"class":"_1bVOdb"}).text=="Isıtma Tipi":
               Isitma_Tipi[ListIndex] += detay.find("div",attrs={"class":"_1bVOdb"}).next_sibling.text   
            if detay.find("div",attrs={"class":"_1bVOdb"}).text=="Krediye Uygunluk":
               Krediye_Uygunluk[ListIndex] += detay.find("div",attrs={"class":"_1bVOdb"}).next_sibling.text
            if detay.find("div",attrs={"class":"_1bVOdb"}).text=="Binanın Yaşı":
               Binanin_Yasi[ListIndex] += detay.find("div",attrs={"class":"_1bVOdb"}).next_sibling.text
            if detay.find("div",attrs={"class":"_1bVOdb"}).text=="Binanın Kat Sayısı":
               Binanin_Kat_Sayisi[ListIndex] += detay.find("div",attrs={"class":"_1bVOdb"}).next_sibling.text
        ListIndex +=1
```

```python
for i in range(0,1054):
    print(fiyatlar[i]+" -*- "+Net_Metrekare[i]+" -*- "+Oda_Sayisi[i]+" -*- "+Isitma_Tipi[i]+" -*- "+Krediye_Uygunluk[i]+" -*- "+Binanin_Yasi[i]+" -*- "+Binanin_Kat_Sayisi[i])
```
